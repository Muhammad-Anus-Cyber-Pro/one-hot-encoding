### **One Hot Encoding**
---

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
df["brand"].value_counts().head(10)
```

### **OneHotEncoding Using Pandas**
---

```python
pd.get_dummies(df,columns=['fuel','owner'],dtype=int)
```

### **K-1 OneHotEncoding**
---

```python
pd.get_dummies(df,columns=["fuel","owner"],drop_first=True)
```

### **OneHotEncoding using Sklearn**
---

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first",sparse_output=False,dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[["fuel","owner"]])
X_test_new = ohe.transform(X_test[["fuel","owner"]])
```

```python
np.hstack((X_train[["brand","km_driven"]].values,X_train_new))
```

### **OneHotEncoding with Top Categories**
---

```python
counts = df["brand"].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
repl
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'),dtype=int).sample(5)
```
